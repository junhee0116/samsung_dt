# 1. 파이썬 파일 입출력

> **파일 입출력 순서**
* 파일을 연다.
* 파일의 내용을 읽거나 쓴다.
* 파일을 닫는다.

> **함수 open()**
* 파이썬에서 파일 입출력에 사용하는 기본 함수
* 프로그램과 파일을 연결하는 내장함수
* 파일객체 = open('파일명', 모드)

> **모드**
* 파일 입출력에서는 **읽기 모드**, **쓰기 모드**, **추가 모드**가 있음.

| 기호 | 모드(용도) | 설명 |
|:--:|:--:|:--:|
|r|읽기 모드|이미 만들어져 있는 파일의 내용을 읽을 때 사용
|w|쓰기 모드|파일을 새로 만들고 내용을 쓸 때 사용
|a|추가 모드|이미 만들어져 있는 파일의 내용 끝에 새로운 내용을 추가할 때 사용



> **파일 포인터**
* **파일 포인터**(file pointer)란 파일을 가르킴.
* 어떤 파일을 연 후, 이름표를 달아 해당 파일에 쓰거나, 읽거나, 추가하는데 활용함.
* 관행에 따라 보통 fp라는 이름을 사용함.

### 1.1.2 파일 입출력 "쓰기" 연습문제
* 문제1) 사용자로부터 텍스트를 입력받아 "output.txt" 파일에 저장하는 파이썬 프로그램을 작성하십시오.
* 문제2) 1부터 10까지의 숫자를 "numbers.txt" 파일에 한 줄씩 *저장하는* 파이썬 프로그램을 작성하십시오.
* 문제3) 주어진 리스트를 "list.txt" 파일에 각 요소를 한 줄씩 저장하는 파이썬 프로그램을 작성하십시오.

```python
# 1)
f = open('output.txt', 'w')

while True:
  a = input()
  if a == '#':
    break
  f.write(f'{a}\n')

f.close()
```

```python
# 2)
f = open('numbers.txt', 'w')

for i in range(1, 11):
  if (i == 1):
      f.write(str(i))
  else:
    f.write(f'\n{str(i)}')

f.close()
```

```python
# 3)
fruits = ["Apple", "Banana", "Orange", "Grape", "Pineapple", "Strawberry", "Blueberry", "Mango", "Watermelon", "Cherry"]

f = open('list.txt', 'w')

for i in range(len(fruits)):
  if (i == 0):
      f.write(fruits[i])
  else:
    f.write(f'\n{fruits[i]}')

f.close()
```

### 1.2. 파일 읽기


### 1.2.1. 파일 읽기 예시

> **read()**
* 파일의 전체 내용을 하나의 문자열로 읽어오기

```python
# read() 사용
# 파일 안에 있는 모든 내용을 하나의 문자열로 만들어서 return 값으로 전달해 줌.

f = open('myfile.txt', 'r')

text = f.read()

print(text)
# print(text.split('\n'))
f.close()
```

```python
# 파일의 라인별로 내용 읽기

f = open('myfile.txt', 'r')

for line in f:
    print(line)

f.close()
```

> **readline()**
* 파일의 전체 내용을 한 줄씩 읽어오기
* 다음 줄을 읽으려면 다시 호출해 줘야 함.
* 문서가 아주 클 경우 사용하면 유용함
* 인수로 읽을 문자 수 지정 가능

```python
# readline() 사용

f = open('myfile.txt', 'r')

for _ in range(5):
    print(f.readline())

f.close()
```

> **readlines()**
* 파일의 전체 내용을 한 번에 리스트로 읽어옴.
* 줄바꿈(\n) 기준으로 문자열을 나눔.
* string.split('\n')
* 파일이 크지 않고 한 번에 읽어도 메모리에 부담이 없는 경우 readlines()를 사용하면 편리함.

```python
# readlines() 사용

f = open('myfile.txt', 'r')

str_list = f.readlines()
print(str_list)

# 추가 설명
print(list(map(lambda x: x.strip(), str_list)))

f.close()
```

```python
# string.split('\n')

f = open('myfile.txt', 'r')
texts = f.read()

result_list = texts.split('\n')
print(result_list)
```

### 1.4.2. 파일 입출력 추가 연습문제
* 문제9) "output.txt" 파일의 내용을 읽어와 각 줄의 길이를 출력하는 파이썬 프로그램을 작성하십시오. 파일을 열 때 with문을 사용하십시오.
* 문제10) 사용자로부터 파일 이름을 입력받고 해당 파일의 내용을 화면에 출력하는 파이썬 프로그램을 작성하십시오. 사용자가 존재하지 않는 파일을 입력할 경우 에러 메시지를 출력하십시오.
* 문제11) "numbers.txt" 파일의 내용을 읽어와 각 줄의 숫자를 합산하여 출력하는 파이썬 프로그램을 작성하십시오. 파일을 열 때 with문을 사용하십시오.

```python
# 9)
with open('output.txt', 'r') as f:
  dic = {}
  text = f.readlines()
  text_len = list(map(lambda x: x.strip(), text))
  for i in range(len(text_len)):
    dic[(i + 1)] = len(text_len[i])

print(dic)
```

```python
# 10)
filename = input('file name: ')
filetype = input('file type: ')

try:
  with open(filename + '.' + filetype, 'r') as f:
    text = f.read()
    print(text)
except FileNotFoundError as e:
  print(e)
```

```python
# 11)
with open('numbers.txt', 'r') as f:
  number = f.read()
  number = number.split('\n')
  number = list(map(lambda x: int(x), number))
  print(np.sum(number))
```

## 1.5. 파일 인코딩과 디코딩

> 파일 인코딩과 디코딩
* 파일의 내용은 미리 정의된 특정 문자열 집합을 사용하여 저장됨(인코딩).
* 파일을 읽을 때는 어떤 문자열 집합으로 읽을 것인지를 알려주어야 함(디코딩).
* 크게 ANSI 또는 UTF-8를 사용
* ANSI로 나올 경우에는 ‘euc-kr’, UTF-8로 나올 경우에는 ‘utf-8’를 사용함.
* 한글 인코딩
 - euc-kr(메모장에서 ANSI로 표현)
 - cp949(=MS949, euc-kr의 확장판)
 - utf-8(ascii)
