```python
import pandas as pd

# 행 단위로 데이터 입력
# 열 2개를 가진 데이터프레임 생성
df = pd.DataFrame([[1, 2], [3, 4], [5, 6]])
df
```

```python
# 행 단위로 데이터 입력
# 열 이름 지정
df = pd.DataFrame([[1, 2], [3, 4], [5, 6]], columns = ['A', 'B'])
df
```

```python
# 행 단위로 데이터 입력
# 인덱스 지정
df = pd.DataFrame([[1, 2], [3, 4], [5, 6]], index = ['A', 'B', 'C'])
df
```

```python
# 행 단위로 데이터 입력
# 딕셔너리를 이용해 데이터프레임 생성
# 딕셔너리는 key: value
dic = [{'A': 1, 'B': 2}, {'A': 3, 'B': 4}, {'A': 5, 'B': 6}]
df = pd.DataFrame(dic)
df
```

```python
# 열 단위로 데이터 입력
# 딕셔너리를 이용해 데이터프레임 생성
dic = {'A': [1, 3, 5], 'B': [2, 4, 6]}
df = pd.DataFrame(dic)
df
```
