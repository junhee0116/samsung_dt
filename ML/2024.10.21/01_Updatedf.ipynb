```python
df['화재크기분류'] = df['재산피해소계'].apply(lambda x: '소형' if x <= 1000 else '대형')

df.head(2)
```

```python
# 한 행씩 함수를 호출
def fire_classification(row):
    if(row['재산피해소계'] < 500):
        return '소형'
    elif(row['재산피해소계'] < 10000):
        return '중형'
    else:
        return '대형'
    
# axis = 1: 행, axis = 0: 열
df['화재크기분류'] = df.apply(fire_classification, axis = 1)

df.head(2)
```

```python
# 중형화재 데이터를 필터하여 출력
# 1. 중형화재로 필터
df[df['화재크기분류'] == '중형'].head(2)

# 2. 재산피해액으로 필터
cond1 = df['재산피해소계'] >= 500
cond2 = df['재산피해소계'] < 10000
df[cond1 & cond2].head(2)
```

```python
import numpy as np
```

```python
# chatgpt
# python에서 pandas로 만든 데이터프레임에는 재산피해소계라는 열이 있다.
# 이 열을 기반으로 화재크기분류라는 열을 만들고 싶은데 그 규칙은 아래와 같다
# 이 작업을 수행할 수 있는 코드를 작성해줘
# --------------------- 규칙 ---------------------
# 소형: 재산피해소계 < 500
# 중형: 500 <= 재산피해소계 < 10000
# 대형: 10000 <= 재산피해소계

conditions = [df['재산피해소계'] < 500, (df['재산피해소계'] >= 500) & (df['재산피해소계'] < 10000), df['재산피해소계'] >= 10000]

choices = ['소형', '중형', '대형']

df['화재크기분류'] = np.select(conditions, choices, default = '')

df.head()
```

```python
# 데이터 수직 결합
df_2018 = pd.read_csv('../data/2018년화재.csv', encoding = 'cp949')
df_2019 = pd.read_csv('../data/2019년화재.csv', encoding = 'cp949')

print('df_2018.shape: {}, df_2019.shape:{}'.format(df_2018.shape, df_2019.shape))

df = pd.concat([df_2018, df_2019])

print('df.shape:', df.shape)
```
