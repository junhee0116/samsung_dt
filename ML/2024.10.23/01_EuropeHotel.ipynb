```python
# 강사님 풀이
import pandas as pd

df = pd.read_csv('../data/Europe Hotel Booking Satisfaction Score.csv')

df.info()
```

```python
# 함수화
from sklearn.preprocessing import LabelEncoder

def preprocess(df, label_column, drop_columns = None, onehot_columns = None):
    if(drop_columns):
        df.drop(drop_columns, axis = 1, inplace = True)
    
    encoder = LabelEncoder()
    encoder.fit(df[label_column])
    df[label_column] = encoder.transform(df[label_column])

    df = pd.get_dummies(df, columns = onehot_columns)

    X = df.drop(label_column, axis = 1)
    y = df[label_column]

    return X, y

X, y = preprocess(df, 'satisfaction', ['id'], ['Gender', 'purpose_of_travel', 'Type of Travel', 'Type Of Booking'])
```

```python
# id열 제거
df.drop('id', axis = 1, inplace = True, errors = 'ignore')
```

```python
# satisfaction 열 레이블 인코딩
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['satisfaction'] = encoder.fit_transform(df['satisfaction'])

df.head(1)
```

```python
# 원-핫 인코딩
df = pd.get_dummies(df, columns = ['Gender', 'purpose_of_travel', 'Type of Travel', 'Type Of Booking'])
```

```python
# X, y로 데이터 분할
X = df.drop('satisfaction', axis = 1)
y = df['satisfaction']
```

```python
# 표준화 혹은 정규화
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X)
scaled_X = scaler.transform(X)
X = pd.DataFrame(scaled_X, columns = X.columns)

X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, stratify = y, random_state = 121)

clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

pred = clf.predict(X_test)

print('score: {:.4f}'.format(accuracy_score(y_test, pred)))

# StandardScaler: 0.8416
# MinMaxScaler: 0.8415
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, pred)
```

```python
# 정밀도
from sklearn.metrics import precision_score

round(precision_score(y_test, pred), 4)
```

```python
# 재현율
from sklearn.metrics import recall_score

round(recall_score(y_test, pred), 4)
```

```python
# Auc
from sklearn.metrics import roc_auc_score

pred = clf.predict_proba(X_test)

print(roc_auc_score(y_test, pred[:, -1]))

pd.DataFrame(pred)
```
