```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../data/Europe Hotel Booking Satisfaction Score.csv')

df.head()
```

```python
df.info()
```

```python
df['Age'].describe()
```

```python
df = df.drop('id', axis = 1, errors = 'ignore')

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder.fit(df['satisfaction'])
df['satisfaction'] = encoder.transform(df['satisfaction'])

df.head()
```

```python
df_dummies = pd.get_dummies(df[['Gender', 'purpose_of_travel', 'Type of Travel', 'Type Of Booking']])

df = df.drop(['Gender', 'purpose_of_travel', 'Type of Travel', 'Type Of Booking'], axis = 1, errors = 'ignore')

df.head()
```

```python
# MinMaxScaler
from sklearn.preprocessing import MinMaxScaler, StandardScaler

y = df['satisfaction']
df = df.drop('satisfaction', axis = 1, errors = 'ignore')
col = df.columns

MinMax_scaler = MinMaxScaler()
MinMax_scaler.fit(df)
Standard_scaler = StandardScaler()
Standard_scaler.fit(df)

MinMax_df = pd.DataFrame(MinMax_scaler.transform(df), columns = col)
Standard_df = pd.DataFrame(Standard_scaler.transform(df), columns = col)

MinMax_df = pd.concat([MinMax_df, df_dummies], axis = 1)
Standard_df = pd.concat([Standard_df, df_dummies], axis = 1)
df = pd.concat([df, df_dummies], axis = 1)
```

```python
MinMax_df.head()
```

```python
Standard_df.head()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(MinMax_df, y, test_size = .3, stratify = y)

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

MinMax_pred = clf.predict(X_test)
MinMax_score = accuracy_score(y_test, MinMax_pred)

X_train, X_test, y_train, y_test = train_test_split(Standard_df, y, test_size = .3, stratify = y)

clf.fit(X_train, y_train)

Standard_pred = clf.predict(X_test)
Standard_score = accuracy_score(y_test, Standard_pred)

X_train, X_test, y_train, y_test = train_test_split(df, y, test_size = .3, stratify = y)

clf.fit(X_train, y_train)

pred = clf.predict(X_test)
score = accuracy_score(y_test, pred)

print('Basic: {:.4f}, MinMax: {:.4f}, Standard: {:.4f}'.format(score, MinMax_score, Standard_score))
```

```python
from sklearn.model_selection import cross_val_score

print('Basic: {:.4f}'.format(cross_val_score(clf, df, y, cv = 3).mean()))
print('MinMax: {:.4f}'.format(cross_val_score(clf, MinMax_df, y, cv = 3).mean()))
print('Standard: {:.4f}'.format(cross_val_score(clf, Standard_df, y, cv = 3).mean()))
```

```python
# 강사님 풀이
import pandas as pd

df = pd.read_csv('../data/Europe Hotel Booking Satisfaction Score.csv')

df.info()
```

```python
# 함수화
from sklearn.preprocessing import LabelEncoder

def preprocess(df, label_column, drop_columns = None, onehot_columns = None):
    if(drop_columns):
        df.drop(drop_columns, axis = 1, inplace = True)
    
    encoder = LabelEncoder()
    encoder.fit(df[label_column])
    df[label_column] = encoder.transform(df[label_column])

    df = pd.get_dummies(df, columns = onehot_columns)

    X = df.drop(label_column, axis = 1)
    y = df[label_column]

    return X, y

X, y = preprocess(df, 'satisfaction', ['id'], ['Gender', 'purpose_of_travel', 'Type of Travel', 'Type Of Booking'])
```

```python
# id열 제거
df.drop('id', axis = 1, inplace = True, errors = 'ignore')
```

```python
# satisfaction 열 레이블 인코딩
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['satisfaction'] = encoder.fit_transform(df['satisfaction'])

df.head(1)
```

```python
# 원-핫 인코딩
df = pd.get_dummies(df, columns = ['Gender', 'purpose_of_travel', 'Type of Travel', 'Type Of Booking'])
```

```python
# X, y로 데이터 분할
X = df.drop('satisfaction', axis = 1)
y = df['satisfaction']
```

```python
# 표준화 혹은 정규화
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X)
scaled_X = scaler.transform(X)
X = pd.DataFrame(scaled_X, columns = X.columns)

X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, stratify = y, random_state = 121)

clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

pred = clf.predict(X_test)

print('score: {:.4f}'.format(accuracy_score(y_test, pred)))

# StandardScaler: 0.8416
# MinMaxScaler: 0.8415
```
